```python
#list the top 15 blue ones
top_blue_percentages = df.sort_values(by='blue_percentage', ascending=False).head(15)
top_blue_filenames = top_blue_percentages['filename'].tolist()
top_blue_filenames
```

```python
#list the top 15 blacks
df.sort_values(by='black_percentage', ascending=False).head(15)
top_black_percentages = df.sort_values(by='black_percentage', ascending=False).head(15)
top_black_filenames = top_black_percentages['filename'].tolist()
top_black_filenames
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def show_image(filenames: list):
    """function for displaying images"""
    
    # Creating a list of paths to images
    image_paths = [ f"{output_folder}/{file}" for file in filenames]

    # Creating a shape and axes to display images
    fig, axes = plt.subplots(5, 1, figsize=(25, 15))

    # We go through each image and display it
    for ax, img_path in zip(axes, image_paths):
        img = mpimg.imread(img_path)
        ax.set_title(img_path[7:])
        ax.imshow(img)
        ax.axis('off')  # Отключаем оси

    plt.show()
```

```python
#show the top 5 blue ones
show_image(top_blue_filenames)
```

```python
#show the top 5 blacks
show_image(top_black_filenames)
```
